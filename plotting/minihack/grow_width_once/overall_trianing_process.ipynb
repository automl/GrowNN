```python
from utils.plotting import get_logtable, set_rc_params, select_incumbents
import pandas as pd
import seaborn as sns
from utils.plotting import training_process_style
import matplotlib.pyplot as plt
from typing import List

set_rc_params()
```

```python
database_name = "fehring_growing_nn"
```

```python
# get data for fixed noise 0.1
smac_callbacks_01 = get_logtable(database_name=database_name, table_name="net2deeper_grow_width_once", logtable_name="smac_callbacks")
smac_callbacks_01 = select_incumbents(smac_callbacks_01 )
smac_callbacks_01
```

```python
# Get data for fixed noise 0.2
smac_callbacks = get_logtable(database_name=database_name, table_name="net2deeper_grow_width_once_new", logtable_name="smac_callbacks")
smac_callbacks_02 = smac_callbacks[smac_callbacks["experiment_id"] == 2]
smac_callbacks_02 = select_incumbents(smac_callbacks_02)
smac_callbacks_02
```

```python
# Get data for optimized noise
smac_callbacks = get_logtable(database_name=database_name, table_name="net2deeper_grow_width_once_new", logtable_name="smac_callbacks")
smac_callbacks_optimized = smac_callbacks[smac_callbacks["experiment_id"] == 5]
smac_callbacks_optimized = select_incumbents(smac_callbacks_optimized)
smac_callbacks_optimized
```

```python
# add to all plots final incumbent for 50 trials
min_01 = smac_callbacks_01["cost"].min()
min_02 = smac_callbacks_02["cost"].min()
min_optimized = smac_callbacks_optimized["cost"].min()

to_concat_01 = pd.DataFrame({"cost": min_01, "trial_number": 50}, index=[0])
to_concat_02 = pd.DataFrame({"cost": min_02, "trial_number": 50}, index=[0])
to_concat_optimized = pd.DataFrame({"cost": min_optimized, "trial_number": 50}, index=[0])

smac_callbacks_01 = pd.concat([smac_callbacks_01, to_concat_01])
smac_callbacks_02 =pd.concat([smac_callbacks_02, to_concat_02])
smac_callbacks_optimized = pd.concat([smac_callbacks_optimized, to_concat_optimized])
```

```python
training_process_style()
sns.lineplot(data=smac_callbacks_01, x="trial_number", y="cost", label="Net2Wider Noise 0.1", drawstyle="steps-post")
sns.lineplot(data=smac_callbacks_02, x="trial_number", y="cost", label="Net2Wider Noise 0.2", drawstyle="steps-post")
sns.lineplot(data=smac_callbacks_optimized, x="trial_number", y="cost", label="Net2Wider Optimized Noise", drawstyle="steps-post")

plt.title("Optimization Process 10x10 Random", fontsize=18, fontweight="bold")
plt.xlabel("Trials", fontsize=14)
plt.ylabel("Cost", fontsize=14)
plt.legend(title="Model Type", fontsize=12, title_fontsize=14, loc="center", bbox_to_anchor=(0.5, -0.16), ncol=3)
plt.grid(True, linestyle="--", alpha=0.7)
plt.savefig("plotting/minihack/grow_width_once/grow_width_once_training_process", bbox_inches="tight")
```
