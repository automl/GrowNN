# Net2DeeperNet Depth 2 Random

```python
base_path = "fehring_growing_nn_new_seeded/net2deeper_budget200_final/4/(32, 0.2, None, 0.0005309372850206717, 0.95, 0.00020422562780736536, 0.5, 10, 2048, False, 0.5, 512, 1, 512, 2)/2/0/model.zip"
data, params, pytorch_variables = load_from_zip_file(base_path) 
# loda 
```

```python
params["policy"].keys()
```

```python
# First layer weights
first_layer_weights = params["policy"]["features_extractor.linear_layers.sequential_container.0.weight"]
first_layer_weights = np.abs(first_layer_weights)
# Select only 100 x 100 weights
first_layer_weights = first_layer_weights[:50, :50]
# Second layer weights
second_layer_weights = params["policy"]["features_extractor.linear_layers.sequential_container.2.weight"]
second_layer_weights = np.abs(second_layer_weights)
# Select only 100 x 100 weights
second_layer_weights = second_layer_weights[:50, :50]
```

```python
# Subplot with 3 figures horically
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
# First layer weights
axs[0].imshow(first_layer_weights, cmap='viridis', vmin=0, vmax=1.3)
axs[0].set_title('Initial Layer Weights', fontsize=17)
# Second layer weights
im3 = axs[1].imshow(second_layer_weights, cmap='viridis', vmin=0, vmax=1.3)
axs[1].set_title('Layer two Weights', fontsize=17)
# Add colorbar for axs[2]
fig.colorbar(im3, ax=axs, orientation='vertical', fraction=0.02, pad=0.04)
plt.suptitle('MiniHack Room Random Net2Deeper Depth 2', fontsize=19, fontweight='bold')
plt.savefig('plotting/network_weights/minihack/net2deeper/random_2.png', bbox_inches='tight')
```
