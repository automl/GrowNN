```python
import numpy as np
import matplotlib.pyplot as plt
from torch import nn
import torch
from typing import Dict
import os
```

# Baselines

```python
base_path = "plotting/network_weights/random_and_identiy_baselines"
```

## 50 x 50

## 50 x 50 matrix

```python
random_layer = nn.Linear(50, 50)
weights = random_layer.weight.detach().numpy()
plt.imshow(weights, cmap="viridis", vmin=0, vmax=1.3)
plt.colorbar()
plt.title('Random Initailisation, 50 x 50', fontweight='bold')
plt.savefig(os.path.join(base_path, "random_initialisation_50x50"))
plt.close()
```

```python
identity_lienar = nn.Linear(50, 50)
identity_lienar.weight.data = torch.eye(50)
identity_matrix = identity_lienar.weight.detach().numpy()
plt.imshow(identity_matrix, cmap='viridis', vmin=0, vmax=1.3)
plt.colorbar()
plt.title('Identity Matrix, 50 x 50', fontweight='bold')
plt.savefig(os.path.join(base_path, "identity_matrix_50x50.png"))
```
