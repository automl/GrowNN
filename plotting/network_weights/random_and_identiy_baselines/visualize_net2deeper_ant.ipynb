```python
import numpy as np
import matplotlib.pyplot as plt
from torch import nn
import torch
from typing import Dict
import os
```

# Baselines

```python
base_path = "plotting/network_weights/random_and_identiy_baselines"
```

## 512 x 512

```python
random_layer = nn.Linear(512, 512)
matrix = random_layer.weight.detach().numpy()

plt.imshow(matrix, cmap='viridis', vmin=0,) 
plt.colorbar()
plt.title('Random Initailisation, 512 x 512', fontweight='bold')
plt.savefig(os.path.join(base_path, "random_initialisation_512x512.png"))
plt.close()
```

```python
identity_lienar = nn.Linear(512, 512)
identity_lienar.weight.data = torch.eye(512
                                        )
identity_matrix = identity_lienar.weight.detach().numpy()
plt.imshow(identity_matrix, cmap='viridis', vmin=0, vmax=1.3)
plt.colorbar()
plt.title('Identity Matrix, 512 x 512', fontweight='bold') 
plt.savefig(os.path.join(base_path, "indentity_matrix_512x512.png"))
plt.close()
```

## 64 x 64 matrix

```python
random_layer = nn.Linear(64, 64)
weights = random_layer.weight.detach().numpy()
plt.imshow(weights, cmap="viridis", vmin=0, vmax=1.3)
plt.colorbar()
plt.title('Random Initailisation, 64 x 64', fontweight='bold')
plt.savefig(os.path.join(base_path, "random_initialisation_64x64"))
plt.close()
```

```python
identity_lienar = nn.Linear(64, 64)
identity_lienar.weight.data = torch.eye(64)
identity_matrix = identity_lienar.weight.detach().numpy()
plt.imshow(identity_matrix, cmap='viridis', vmin=0, vmax=1.3)
plt.colorbar()
plt.title('Identity Matrix, 64 x 64', fontweight='bold')
plt.savefig(os.path.join(base_path, "identity_matrix_64x64.png"))
```
